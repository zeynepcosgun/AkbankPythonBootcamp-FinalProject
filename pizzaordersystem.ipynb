```python
import csv
import datetime

with open('Menu.txt', 'w') as f:
    f.write('* Lütfen Bir Pizza Tabanı Seçiniz:\n'
            '1: Klasik Pizza\n'
            '2: Margarita Pizza\n'
            '3: Türk Pizza\n'
            '4: Sade Pizza\n'
            '* ve seçeceğiniz sos:\n'
            '10: Zeytin\n'
            '11: Mantar\n'
            '12: Keçi Peyniri\n'
            '13: Domates\n'
            '* Teşekkür ederiz!\n')


class Pizza:
    def __init__(self, description, cost):
        self.description = description
        self.cost = cost
        
    def get_description(self):
        return self.description
    
    def get_cost(self):
        return self.cost


class KlasikPizza(Pizza):
    def __init__(self):
        super().__init__('Klasik Pizza', 100)

class MargaritaPizza(Pizza):
    def __init__(self):
        super().__init__('Margarita Pizza', 120)

class TurkPizza(Pizza):
    def __init__(self):
        super().__init__('Türk Pizza', 110)

class SadePizza(Pizza):
    def __init__(self):
        super().__init__('Sade Pizza', 95)


class Decorator(Pizza):
    def __init__(self, component):
        self.component = component
        
    def get_cost(self):
        return self.component.get_cost() + Pizza.get_cost(self)
    
    def get_description(self):
        return self.component.get_description() + ' ' + Pizza.get_description(self)


class Zeytin(Decorator):
    def __init__(self, component):
        super().__init__(component)
        self.description = 'Zeytin'
        self.cost = 10
        
class Mantar(Decorator):
    def __init__(self, component):
        super().__init__(component)
        self.description = 'Mantar'
        self.cost = 18
        
class KeciPeyniri(Decorator):
    def __init__(self, component):
        super().__init__(component)
        self.description = 'Keçi Peyniri'
        self.cost = 24

class Domates(Decorator):
    def __init__(self, component):
        super().__init__(component)
        self.description = 'Domates'
        self.cost = 10


def main():
    
    with open("Menu.txt", "r") as f:
        menu = f.read()

  
    print(menu)

   
    pizza_choice = input("Lütfen bir pizza tabanı seçin (1-4): ")
    if pizza_choice == "1":
        pizza = KlasikPizza()
    elif pizza_choice == "2":
        pizza = MargaritaPizza()
    elif pizza_choice == "3":
        pizza = TurkPizza()
    elif pizza_choice == "4":
        pizza = SadePizza()
    else:
        print("Hatalı giriş yaptınız.")
        return


    sos_choice = input("Lütfen bir sos seçin (10-13): ")
    if sos_choice == "10":
        sos = Zeytin(pizza)
    elif sos_choice == "11":
        sos = Mantar(pizza)
    elif sos_choice == "12":
        sos = KeciPeyniri(pizza)
    elif sos_choice == "13":
        sos = Domates(pizza)        
    else:
        print("Hatalı giriş yaptınız.")
        return

    
    isim = input("İsim: ")
    tc_kimlik_no = input("TC Kimlik No: ")
    kredi_karti_no = input("Kredi Kartı No: ")
    kredi_karti_sifre = input("Kredi Kartı Şifresi: ")

    
    now = datetime.datetime.now()
    order = [isim, tc_kimlik_no, kredi_karti_no, sos.get_description(), str(now), kredi_karti_sifre]
    with open("Orders_Database.csv", "a") as f:
        writer = csv.writer(f)
        writer.writerow(order)

    
    toplam_fiyat = sos.get_cost()
    print("Toplam Fiyat: ", toplam_fiyat)

if __name__ == "__main__":
  main()
```
